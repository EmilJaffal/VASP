```python
import os
import glob
import matplotlib.pyplot as plt

def readDOSCAR(filename, folder_name, legend_labels):
    doscar = open(filename, "r")

    doscar.readline()  # Skip file header
    doscar.readline()
    doscar.readline()
    doscar.readline()

    name = doscar.readline().rstrip()

    first = doscar.readline()
    (emax, emin, ndos, efermi, weight) = first.split()

    es = []
    tds = []
    ids = []

    print("# Filename:", os.path.basename(filename))
    print("# Folder:", folder_name)
    print("# Max Energy: %.4f eV" % float(emax))
    print("# Min Energy: %.4f eV" % float(emin))
    print("# Fermi Energy: %.4f eV" % float(efermi))
    print("#", ndos, "points calculated")
    print()

    for i in range(int(ndos)):
        line = doscar.readline()
        (energy, tdos, idos) = line.split()

        try:
            energy = float(energy)
            tdos = float(tdos)
            idos = float(idos)
        except ValueError:
            continue

        tds.append(tdos)
        es.append(energy - float(efermi))
        ids.append(idos)

    plt.subplot(1, 2, 1)
    plt.plot(tds, es, '-', label=folder_name)  # Use folder name as the label
    plt.xlabel('States')
    plt.ylim(-12, 27)  # Set the y-axis limits... adjust as needed
    plt.ylabel('Energy (eV)')  # Change y-label
    plt.title('Total DOS')

    plt.subplot(1, 2, 2)
    plt.plot(ids, es, '-')
    plt.xlabel('States')
    plt.ylim(-12, 27)
    plt.ylabel('Energy (eV)')  # Change y-label
    plt.title('Integrated DOS')

    legend_labels.append(folder_name)  # Append folder name for legend

    plt.subplots_adjust(wspace=0.4)  # Adjust the space between subplots

    plt.suptitle("ReCo2In Series DOS", fontsize=16)  # Add a title to the entire figure

def main(directory):

    plt.figure('Density of States', figsize=(15, 9))  # Increase the figure size

    # Use glob to find all files ending with "DOSCAR" in the specified directory
    doscar_files = glob.glob(os.path.join(directory, '**', '*DOSCAR'), recursive=True)
    doscar_files = [f for f in doscar_files if "ispin=2" not in f and "Not_Using" not in f]

    if not doscar_files:
        print("No DOSCAR files found in the specified directory or its subdirectories.")
        return

    legend_labels = []  # Store legend labels

    for filename in doscar_files:
        folder_name = os.path.basename(os.path.dirname(filename))  # Extract parent folder name
        readDOSCAR(filename, folder_name, legend_labels)
        print()  # Add an empty line between each block of information

    # Add legend based on folder names
    plt.subplot(1, 2, 1)
    plt.legend(legend_labels, loc='upper right')

    # Save the plot as PNG
    output_filename = os.path.join(os.path.expanduser('~'), 'Desktop/Molecules/Graphs/DOS', 'Myriad.png')
    plt.savefig(output_filename)
    plt.show()  # Display the plot
    plt.close()  # Close the figure to free up memory

main('Desktop/Molecules')
```

# DOS spin up/down & IDOS spin up/down - Myriad:

```python
import os
import glob
import matplotlib.pyplot as plt

def readDOSCAR(filename, folder_name, all_tds_up, all_tds_down, all_ids_up, all_ids_down, parent_folder):
    doscar = open(filename, "r")

    doscar.readline()  # Skip file header
    doscar.readline()
    doscar.readline()
    doscar.readline()

    doscar.readline()  # Skip system name

    first = doscar.readline()
    (emax, emin, ndos, efermi, weight) = first.split()

    es = []
    tds_up = []
    tds_down = []
    ids_up = []
    ids_down = []

    for i in range(int(ndos)):
        line = doscar.readline()
        (energy, tdos_up, tdos_down, idos_up, idos_down) = line.split()

        try:
            energy = float(energy)
            tdos_up = float(tdos_up)
            tdos_down = float(tdos_down)
            idos_up = float(idos_up)
            idos_down = float(idos_down)
        except ValueError:
            continue

        es.append(energy - float(efermi))
        tds_up.append(tdos_up)
        tds_down.append(tdos_down)
        ids_up.append(idos_up)
        ids_down.append(idos_down)

    all_tds_up.append(tds_up)
    all_tds_down.append(tds_down)
    all_ids_up.append(ids_up)
    all_ids_down.append(ids_down)

    return es, parent_folder

def plot_combined_graph(ax, es, all_data, ylabel, title, legend_labels):
    for i, data in enumerate(all_data):
        ax.plot(data, es, label=legend_labels[i])

    ax.set_xlabel(ylabel)
    ax.set_ylabel(r'Energy (eV)')
    ax.set_title(title)
    ax.legend()
    ax.grid(False)  # Remove grid lines
    ax.set_ylim(-10, 25)  # Set y-axis limits

def main(directory):
    all_tds_up = []
    all_tds_down = []
    all_ids_up = []
    all_ids_down = []
    all_legend_labels = []

    # Use glob to find all files ending with "DOSCAR" in the specified directory
    doscar_files = glob.glob(os.path.join(directory, '**', '*DOSCAR'), recursive=True)
    doscar_files = [f for f in doscar_files if "ispin=2" in f and "Not_Using" not in f]

    if not doscar_files:
        print("No DOSCAR files found in the specified directory or its subdirectories.")
        return

    for filename in doscar_files:
        folder_name = os.path.basename(os.path.dirname(filename))  # Extract parent folder name
        parent_folder = os.path.basename(os.path.dirname(os.path.dirname(filename)))  # Extract parent folder of ispin=2 directory
        all_legend_labels.append(parent_folder)
        es, _ = readDOSCAR(filename, folder_name, all_tds_up, all_tds_down, all_ids_up, all_ids_down, parent_folder)

    fig, axes = plt.subplots(2, 2, figsize=(9, 15))

    # Plot combined graphs for DOS (up), DOS (down), IDOS (up), and IDOS (down)
    plot_combined_graph(axes[0, 0], es, all_tds_up, 'States', 'ReCo2In Series ISPIN=2 TDOS (up)', all_legend_labels)
    plot_combined_graph(axes[0, 1], es, all_tds_down, 'States', 'ReCo2In Series ISPIN=2 TDOS (down)', all_legend_labels)
    plot_combined_graph(axes[1, 0], es, all_ids_up, 'States', 'ReCo2In Series ISPIN=2 IDOS (up)', all_legend_labels)
    plot_combined_graph(axes[1, 1], es, all_ids_down, 'States', 'ReCo2In Series ISPIN=2 IDOS (down)', all_legend_labels)

    plt.tight_layout()
    plt.savefig('Desktop/Molecules/Graphs/ISPIN=2/combined_plots.png')
    plt.show()

main('Desktop/Molecules')
```

# Individual DOS spin up/down & IDOS spin up/down

```python
import os
import glob
import matplotlib.pyplot as plt

def readDOSCAR(filename, folder_name):
    doscar = open(filename, "r")

    doscar.readline()  # Skip file header
    doscar.readline()
    doscar.readline()
    doscar.readline()

    doscar.readline()  # Skip system name

    first = doscar.readline()
    (emax, emin, ndos, efermi, weight) = first.split()

    es = []
    tds_up = []
    tds_down = []
    ids_up = []
    ids_down = []

    print("# Filename:", os.path.basename(filename))
    print("# Folder:", folder_name)
    print("# Max Energy: %.4f eV" % float(emax))
    print("# Min Energy: %.4f eV" % float(emin))
    print("# Fermi Energy: %.4f eV" % float(efermi))
    print("#", ndos, "points calculated")
    print()

    for i in range(int(ndos)):
        line = doscar.readline()
        (energy, tdos_up, tdos_down, idos_up, idos_down) = line.split()

        try:
            energy = float(energy)
            tdos_up = float(tdos_up)
            tdos_down = float(tdos_down)
            idos_up = float(idos_up)
            idos_down = float(idos_down)
        except ValueError:
            continue

        es.append(energy - float(efermi))
        tds_up.append(tdos_up)
        tds_down.append(tdos_down)
        ids_up.append(idos_up)
        ids_down.append(idos_down)

    fig, axs = plt.subplots(1, 2, figsize=(15, 9))

    # Plot DOS(up) and DOS(down)
    axs[0].plot(tds_up, es, '-', label='DOS(up)')  # Plot DOS(up)
    axs[0].plot(tds_down, es, '--', label='DOS(down)')  # Plot DOS(down)
    axs[0].set_xlabel('States')
    axs[0].set_ylabel('Energy (eV)')
    axs[0].set_title(folder_name + " ISPIN=2 DOS")  # Set title to molecule name
    axs[0].legend()

    # Plot IDOS(up) and IDOS(down)
    axs[1].plot(ids_up, es, '-', label='IDOS(up)')  # Plot IDOS(up)
    axs[1].plot(ids_down, es, '--', label='IDOS(down)')  # Plot IDOS(down)
    axs[1].set_xlabel('States')
    axs[1].set_ylabel('Energy (eV)')
    axs[1].set_title(folder_name + " ISPIN=2 IDOS")  # Set title to molecule name
    axs[1].legend()

    # Save the plot as PNG based on folder name
    output_filename = os.path.join(os.path.expanduser('~'), 'Desktop/Molecules/Graphs/ISPIN=2', folder_name + ".png")
    plt.savefig(output_filename)
    plt.show()  # Display the plot
    plt.close()  # Close the figure to free up memory

def main(directory):

    # Use glob to find all files ending with "DOSCAR" in the specified directory
    doscar_files = glob.glob(os.path.join(directory, '**', '*DOSCAR'), recursive=True)
    doscar_files = [f for f in doscar_files if "ispin=2" in f and "Not_Using" not in f]

    if not doscar_files:
        print("No DOSCAR files found in the specified directory or its subdirectories.")
        return

    for filename in doscar_files:
        folder_name = os.path.basename(os.path.dirname(os.path.dirname(filename)))  # Extract parent folder name
        readDOSCAR(filename, folder_name)
        print()  # Add an empty line between each block of information

main('Desktop/Molecules')
```
