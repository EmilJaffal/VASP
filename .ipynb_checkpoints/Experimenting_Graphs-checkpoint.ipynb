```python
import os
from abipy.abilab import LobsterAnalyzer

# Define the path to the Lobster file
filepath = '/Users/emiljaffal/Desktop/Structures/CeCo2In/Lobsterpy/COOPCAR.lobster'

# Check if the file exists
if os.path.exists(filepath):
    # Extract directory path and filename
    dirpath, filename = os.path.split(filepath)
    
    # Initialize the LobsterAnalyzer instance with the directory path and prefix
    lobana = LobsterAnalyzer(dirpath, prefix=filename)

    # Load data from the file
    lobana.read(filepath)
    
    # Print the LobsterAnalyzer object
    print(lobana)

    # Plot COOP + COHP + DOS.
    lobana.plot(title="COOP + COHP + DOS")

    # Plot COHP for all sites in from_site_index and Lobster DOS.
    lobana.plot_coxp_with_dos(from_site_index=[0, 1])

    # Plot orbital projections.
    lobana.plot_coxp_with_dos(from_site_index=[0], with_orbitals=True)

    #lobana.plot_with_ebands(ebands="out_GSR.nc")
else:
    print("File not found:", filepath)
```

```python
#https://abinit.github.io/abipy/gallery/plot_lobster_cohp.html#sphx-glr-gallery-plot-lobster-cohp-py

import os
import abipy.data as abidata
from abipy.abilab import abiopen

dirpath = os.path.join("/Users/emiljaffal/Desktop/Structures/CeCo2In/Lobster")
filename = os.path.join(dirpath, "COOPCAR.lobster")

# Open the COHPCAR.lobster file (same API for COOPCAR.lobster)
coop_file = abiopen(filename)
print(coop_file)

# Plot COHP.
coop_file.plot(title="CeCo2In COOP")

# Plot integrated COHP.
coop_file.plot(what="i", title="CeCo2In integrated COOP")

# Plot total overlap for all sites listed in `from_site_index`
coop_file.plot_site_pairs_total(from_site_index=[0], title="COOP total overlap for site index 0")

# Plot partial crystal orbital projections for all sites listed in `from_site_index`
coop_file.plot_site_pairs_partial(from_site_index=[0], title="COOP with orbital projections from site index 0")
```

```python
import os
import pandas as pd

directory = 'Downloads/CGCNN_CVS_RESULTS'

for i in range(1, 11):
    folder_path = os.path.join(directory, str(i))
    if os.path.exists(folder_path):
        csv_file = os.path.join(folder_path, 'test_results.csv')
        if os.path.exists(csv_file):
            # Read the existing content of the CSV file
            df = pd.read_csv(csv_file)
            # Rename the 'Run' and 'Prediction' columns based on the parent folder name
            df = df.rename(columns={'Run': f'Run_{i}', 'Prediction': f'Prediction_{i}'})
            # Save as new_test_results.csv in the same folder
            new_csv_file = os.path.join(folder_path, 'new_test_results.csv')
            # Write the modified dataframe to CSV file with the headers
            df.to_csv(new_csv_file, index=False)
            print(f"Saved modified dataframe to {new_csv_file}")
        else:
            print(f"test_results.csv not found in folder {i}.")
    else:
        print(f"Folder {i} not found.")
```

```python
import os
import pandas as pd

directory = 'Downloads/CGCNN_CVS_RESULTS'
merged_df = None

for i in range(1, 11):
    folder_path = os.path.join(directory, str(i))
    csv_file = os.path.join(folder_path, 'new_test_results.csv')
    
    if os.path.exists(csv_file):
        # Read the CSV file into a dataframe
        df = pd.read_csv(csv_file)
        # If merged_df is None, assign it the current dataframe; otherwise, merge the dataframes
        if merged_df is None:
            merged_df = df
        else:
            # Merge dataframes based on the 'Entry' column
            merged_df = pd.merge(merged_df, df, on='Entry', how='outer')
    else:
        print(f"new_test_results.csv not found in folder {i}.")

# Save the merged dataframe to a new CSV file
merged_csv_file = os.path.join(directory, 'merged_results.csv')
merged_df.to_csv(merged_csv_file, index=False)
print(f"Saved merged dataframe to {merged_csv_file}")
```
